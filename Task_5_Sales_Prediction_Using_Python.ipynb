### Introduction

Sales prediction is the process of forecasting the quantity of a product that consumers will purchase depending on variables like the advertising budget, the target market, and the platform you are using.

A company that sells goods and services usually needs a data scientist to forecast future sales for each action they take to control the expense of product advertising. So let's get started on the Python machine learning sales prediction job.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
x=df[ ['Unnamed: 0','TV', 'Radio', 'Newspaper']]
y=df[[ 'Sales' ]]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OrdinalEncoder
scaler = StandardScaler()
x= scaler.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=.2,shuffle=True)
```

```python
model = LinearRegression()

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print(f''' The train accuracy : {r2_score(y_train,model.predict(x_train))}
The test accuracy : {r2_score(y_test , y_pred)}''')
```

```python
random_forest = RandomForestRegressor(n_estimators=100)
random_forest.fit(x_train,y_train)
```

```python
random_f_pred = random_forest.predict(x_test)
```

```python
r2_score(y_test,random_f_pred)
```

### Model is good fit on Train dataset & Performance of model on Test data is also good fit

### CONCLUSION
